```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import math
import multiprocessing
from sklearn.metrics import mean_absolute_error

pd.options.display.max_rows = 200

np.set_printoptions(threshold=np.inf)
np.set_printoptions(linewidth=100, suppress=True)
```

```python
so = pd.read_csv("../result/all_pred_testing_ratings.csv")
```

```python
def shuf(series,factor=2):
    z = float(series.max()) - float(series.min())
    a = float(factor)*z/50
    return series.apply(lambda x: x+np.random.uniform(-a,a))
```

```python
hl = np.arange(0.5,5.5,0.5)

plt.figure(figsize=(12,4))
plt.subplot(121)
plt.axhline(y=0, c='k',alpha=0.4)
plt.vlines(0.5,-4.5,0.5,linestyle='--',colors='k',alpha=0.4)
plt.vlines(5,0,5,linestyle='--',colors='k',alpha=0.4)
plt.plot(hl,hl-5,linestyle='--',c='k',alpha=0.4)
plt.plot(hl,hl,linestyle='--',c='k',alpha=0.4)
plt.scatter(shuf(so['pred3']),shuf(so['pred3'])-so['rating'],c='g',marker='.',alpha=0.05)
plt.title('Slope One')
plt.ylabel('Residuals')
plt.xlabel('Predicted values')

plt.subplot(122)
plt.axhline(y=0, c='k',alpha=0.4)
plt.vlines(0.5,-4.5,0.5,linestyle='--',colors='k',alpha=0.4)
plt.vlines(5,0,5,linestyle='--',colors='k',alpha=0.4)
plt.plot(hl,hl-5,linestyle='--',c='k',alpha=0.4)
plt.plot(hl,hl,linestyle='--',c='k',alpha=0.4)
plt.scatter(shuf(so['pred1']),shuf(so['pred1'])-so['rating'],c='g',marker='.',alpha=0.05)
plt.title('Slope One without adjusting steps')
plt.ylabel('Residuals')
plt.xlabel('Predicted values')

plt.savefig('../result/Residuals_Predicted.png',dpi=200)
```
